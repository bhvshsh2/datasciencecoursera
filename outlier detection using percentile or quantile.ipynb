```python
#now calculating percentile of price_per_sq
max_thresold=df['price_per_sq'].quantile(.99)
print('max thresold per sqft',max_thresold)
min_thresold=df['price_per_sq'].quantile(.001)
print('min thresold persqft',min_thresold)
```

```python
#now analizing datapointof  min and max thresold
print('datapoints greater then max thresold  :',len(df[df['price_per_sq']>max_thresold]))
print('datapoints smaller than min thresold   :',len(df[df['price_per_sq']<min_thresold]))
```
