```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style

classes=np.array(['python','r','AI','ML','DS'])
student1=np.array([45,5,30,10,10])
student2=np.array([10,20,30,30,10])
student3=np.array([15,20,10,30,20])
```

```python
plt.bar(classes,student1)
```

```python
plt.bar(classes,student1,color='y',width=.2,edgecolor='m',label ='student 1 of every class',linewidth=5,alpha=.7)#alpha is used opacity 
plt.legend()
```

```python
class_index=np.arange(len(classes))
width=.2
plt.bar(class_index,student1,width,color='y',label='student 1')#alpha is used opacity 

plt.bar(class_index+width,student2,width,color='b',label='student 2')

plt.bar(class_index+width+width,student3,width,color='m',label='student 3')

plt.legend()
plt.xticks(class_index+width,classes,rotation=20)
plt.title('bar chart of Datascience students')
plt.xlabel('subjects')
plt.ylabel('no of students')
```
